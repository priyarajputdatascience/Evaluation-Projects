# Data Modeling:-

# Replacing missing values-

```python
#Filling the values in Workclass, Occupation and Native_country column using the most frequent values:

df['Workclass']=df['Workclass'].replace('?', 'Private')
df['Occupation']=df['Occupation'].replace('?', 'Prof-specialty')
df['Native_country']=df['Native_country'].replace('?', 'United_States')
```

# Catagorical Data Handling:-

```python
df.dtypes
```

```python
for column in df.columns:
    print(column,df[column].nunique())
```

# Apply label encoder:-

```python
#converting data to numerical type using LabelEncoder:
from sklearn.preprocessing import LabelEncoder
label=LabelEncoder()

label.fit(df.Sex.drop_duplicates())
df.Sex= label.transform(df.Sex)

label.fit(df.Income.drop_duplicates())
df.Income= label.transform(df.Income)

df.dtypes
```

# Apply ordinal encoder:-

```python
from sklearn.preprocessing import OrdinalEncoder
```

```python
enc=OrdinalEncoder()
df['Workclass']=enc.fit_transform(df[['Workclass']])
df['Education']=enc.fit_transform(df[['Education']])
df['Marital_status']=enc.fit_transform(df[['Marital_status']])
df['Occupation']=enc.fit_transform(df[['Occupation']])
df['Relationship']=enc.fit_transform(df[['Relationship']])
df['Race']=enc.fit_transform(df[['Race']])
df['Native_country']=enc.fit_transform(df[['Native_country']])

df.dtypes
```

# Skewness Handling:

```python
plt.figure(figsize=(25,20))
for i in enumerate(df_new.columns):
    plt.subplot(8,4,i[0]+1)
    sns.distplot(df_new[i[1]],color='b')
```

* Skewness more than +/-0.5 will be treated.

```python
#Removing skewness of the data:

from sklearn.preprocessing import PowerTransformer
PT=PowerTransformer()
for i in df_new.columns:
    if abs(df_new.loc[:,i].skew())>0.5:
        df_new.loc[:,i]=PT.fit_transform(df_new.loc[:,i].values.reshape(-1,1))
```

# Balancing the target column using SMOTE:-

```python
from collections import Counter
from imblearn.over_sampling import SMOTE
sm=SMOTE()
print('unbalanced data: ',Counter(y))
X_sm,y_sm=sm.fit_resample(X,y)
print('balanced data: ',Counter(y_sm))
```

# Machine Learning model building:

```python
#import libraries:-
from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.ensemble import AdaBoostClassifier
from sklearn.ensemble import GradientBoostingClassifier

from sklearn.svm import SVC
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import accuracy_score,confusion_matrix,classification_report
from sklearn.model_selection import cross_val_score,GridSearchCV
```

```python
#Creating the data sets:

X_train,X_test,y_train,y_test=train_test_split(X_sm,y_sm,test_size=0.30,random_state=42,shuffle=True)
```

# Logistic Regression:-

```python
LR= LogisticRegression()

#training the model:-
LR.fit(X_train,y_train)
pred=LR.predict(X_test)

#Checking Accuracy Score:
acc_score=(accuracy_score(y_test,pred))
print("Accuracy_score:",acc_score)

#confusion matrix:-
print('Confusion matrix:\n',confusion_matrix(y_test,pred))

#classification report:-
class_report=classification_report(y_test,pred)
print('n\Classification Report:\n',class_report)

#cross validation score:-
cv_score=(cross_val_score(LR,X,y,cv=5).mean())
print("Cross Validation Score:",cv_score)

#Result of accuracy minus cv scores:-
result=acc_score - cv_score
print("\n Accuracy Score - cross validation score is",result) 
```
